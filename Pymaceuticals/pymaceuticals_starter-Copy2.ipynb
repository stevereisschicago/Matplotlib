```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen


getmean = clean_results_df.groupby('Drug Regimen').mean()['Tumor Volume (mm3)']
getmedian = clean_results_df.groupby('Drug Regimen').median()['Tumor Volume (mm3)']
getvariance = clean_results_df.groupby('Drug Regimen').var()['Tumor Volume (mm3)']
getsds = clean_results_df.groupby('Drug Regimen').std()['Tumor Volume (mm3)']
getsem = clean_results_df.groupby('Drug Regimen').sem()['Tumor Volume (mm3)']


stat_summary = pd.DataFrame({"Mean Tumor Volume":getmean,
                              "Median Tumor Volume":getmedian,
                              "Tumor Volume Variance":getvariance,
                              "Tumor Volume Std. Dev.":getsds,
                              "Tumor Volume Std. Err.":getsem})

stat_summary
```

```python
# Using the aggregation method, produce the same summary statistics in a single line

stat_line_summary= clean_results_df.groupby("Drug Regimen").agg({"Tumor Volume (mm3)":["mean","median","var","std","sem"]})
stat_line_summary
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen

print(f"Correlation Coefficient Between Mouse Weight and Average Tumor Volume for the Capomulin Regiment is: {round(st.pearsonr(x_values, y_values)[0],2)}")
```

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen

    
mw_slope, mw_int, mw_r, mw_p, mw_std_err = st.linregress(x_values, y_values)
print(f'slope = {mw_slope}')
print(f'intercept = {mw_int}')


```

```python
mw_fit = mw_slope * x_values + mw_int
```

```python
x_values = capomulin_df["Tumor Volume (mm3)"]
y_values = capomulin_df["Weight (g)"]
plt.scatter(x_values, y_values)
plt.plot(x_values, mw_fit, "--")

plt.title("Capomulin Regimen - Tumor Volume vs. Mouse Weight")
plt.xlabel("Tumor Volume")
plt.ylabel("Mouse Weight")
plt.show
```
